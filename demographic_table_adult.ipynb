```python
# DENSITY

def cal_demo(column):
    def add_col(x, df, column):
        return df[df['NONCARB'] == x['NONCARB']][column].sum()
    df[column] = df[column].fillna(-1)
    group_columns = ['NONCARB',column]
    output_columns = [column, '_LLCPWT2']

    t_df = df.groupby(group_columns)[output_columns].agg({ column: 'count', '_LLCPWT2': 'sum'}).rename(columns={column: 'count'}).reset_index()
    n_df['TOTAL'] = df.groupby('NONCARB')[column].sum().reset_index()[column].tolist()
    n_df['WTOTAL'] = df.groupby('NONCARB')['_LLCPWT2'].sum().reset_index()['_LLCPWT2'].tolist()

    t_df = t_df.merge(n_df[['NONCARB', 'TOTAL', 'WTOTAL']], how='left', on='NONCARB')
    t_df['percent'] = (t_df['_LLCPWT2']*100)/t_df['WTOTAL']
    return t_df
```
