```python
# Ever told asthma: No
lifetime_asthma_ques = df1.groupby(['_STATE','ASTHMA3'])['ASTHMA3'].agg(['count']).reset_index()
never_asthma = lifetime_asthma_ques["ASTHMA3"]==2.0
never_asthma_df = lifetime_asthma_ques[never_asthma]
```

```python
fips_codes = "data/us-state-ansi-fips.csv"
fips = pd.read_csv(fips_codes)
fips.head(4)
fips.columns
```

```python
acbs_participating_states = incidence_by_state[['_state']].merge(fips, left_on="_state", right_on=' st')
```

```python
acbs_participating_states
```

```python
f_epa = "data/states_and_counties.csv"
epa_codes = pd.read_csv(f_epa)
epa_codes = epa_codes[epa_codes['State Code']!='CC']
epa_codes['State Code'] = epa_codes['State Code'].apply(int)
```

```python
epa_codes = epa_codes.drop_duplicates(['State Code'], keep='first')
```

```python
epa_codes['State Code'].unique()
```

```python
acbc_states_by_epa_region = acbs_participating_states.merge(epa_codes[['State Code', 'State Name', 'EPA Region']], left_on=" st", right_on='State Code', how='left')
```
