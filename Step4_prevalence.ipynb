```python
import pandas as pd
```

```python
fname = "data/BRFSS/2017/current_adult_prevalance_rate.xlsx"
fregion = "data/states_and_counties.csv"
```

```python
df1 = pd.read_excel(fname)
region_df = pd.read_csv(fregion)

region_df = region_df.drop_duplicates(['State Code'], keep='first')
prevalance_df = pd.merge(df1, region_df, how='inner', left_on='State', right_on='State Abbreviation')
prevalance_df = prevalance_df[['State Code', 'State Name', 'EPA Region', 'Prevalence',
               'Standard', '95% CI¶']]

def get_low_CI(x):
    return float(x.split('–')[0].replace("(", ""))
    
def get_high_CI(x):
    return float(x.split('–')[1].replace(")", ""))

prevalance_df['high_CI'] = prevalance_df['95% CI¶'].apply(get_high_CI)
prevalance_df['low_CI'] = prevalance_df['95% CI¶'].apply(get_low_CI)
```
