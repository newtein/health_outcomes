```python
import pandas as pd
from constants import *
import os
```

```python
def read_data(year, fname):
    df = pd.read_csv(fname, index_col=0)
    zev = df.loc["NONCARB", "Odds Ratio"]
    pov = df.loc["POVERTY", "Odds Ratio"]
    print(year, zev, pov)
    return df
```

```python
for state_code in ZEV_STATES:
    years = list(range(2008, 2020))
    dir_path = "odds_ratio_module/data/{}/{}"
    print(state_code)
    for year in years:
        path = dir_path.format(year, "epa_region_0_CHILD_zev_{}.csv".format(state_code))
        if os.path.exists(path):
            read_data(year, path)
```
