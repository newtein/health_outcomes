```python
import pandas as pd
```

```python
path = 'output_files'
OR_file = "{}/statewise_or_2010_2019.xlsx".format(path)
aOR_file = "{}/statewise_aOR.csv".format(path)

def remove_junk(x):
    x = x.split('\n')[0]
    x = x.split()[1]
    return x

def clean_df(df):
    df['PR'] = df['PR'].apply(remove_junk)
    df['IR'] = df['IR'].apply(remove_junk)
    df['AF'] = df['AF'].apply(remove_junk)
    df['OR'] = df['OddsRatio2']
    drop_col = ["Unnamed: 11", "Unnamed: 12", "Unnamed: 13", "Unnamed: 14", "OddsRatio", "OddsRatio2"]
    df.drop(drop_col, axis=1, inplace=True)
    return df

df_or = pd.DataFrame()
for year in range(2010, 2020):
    tdf = pd.read_excel(OR_file, sheet_name="{}".format(year))
    tdf['year'] = year
    df_or = df_or.append(tdf)
```

```python
df_or = clean_df(df_or)
df_aOR = pd.read_csv(aOR_file)
df = df_or.merge(df_aOR, on=['state_code', 'year'], how='left')
df.to_csv("odds_ratio_module/combined_OR_aOR.csv", index=False)
```

```python
path = 'output_files'
OR_file = "{}/statewise_AF_pm2.5_2013_2019.xlsx".format(path)

def remove_junk(x):
    x = x.split('\n')[0]
    x = x.split()[1]
    return x

def clean_df(df):
    df['PR'] = df['PR'].apply(remove_junk)
    df['IR'] = df['IR'].apply(remove_junk)
    df['AF'] = df['AF'].apply(remove_junk)
    drop_col = ["Unnamed: 11", "Unnamed: 12", "Unnamed: 13", "Unnamed: 14"]
    try:
        df.drop(drop_col, axis=1, inplace=True)
    except:
        pass
    return df

df_or = pd.DataFrame()
for year in range(2013, 2020):
    tdf = pd.read_excel(OR_file, sheet_name="{}".format(year))
    tdf['year'] = year
    df_or = df_or.append(tdf)
    
df_or = clean_df(df_or)
df_or.to_csv("output_files/combined_AF_pm2.5.csv", index=False)
```
