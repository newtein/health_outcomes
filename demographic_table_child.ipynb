```python
fname = "odds_ratio_module/data/BRFSS_CHILD_OR.csv"
df = pd.read_csv(fname)
```

```python
pd.options.display.float_format = '{:.2f}'.format
```

```python
def cal_demo(column):
    def add_col(x, df, column):
        return df[df['NONCARB'] == x['NONCARB']][column].sum()
    df[column] = df[column].fillna(-1)
    group_columns = ['NONCARB',column]
    output_columns = [column, '_CLLCPWT']

    t_df = df.groupby(group_columns)[output_columns].agg({ column: 'count', '_CLLCPWT': 'sum'}).rename(columns={column: 'count'}).reset_index()
    n_df = df.groupby('NONCARB')[column].sum().reset_index()
    n_df['TOTAL'] = n_df[column].tolist()
    n_df['WTOTAL'] = df.groupby('NONCARB')['_CLLCPWT'].sum().reset_index()['_CLLCPWT'].tolist()

    t_df = t_df.merge(n_df[['NONCARB', 'TOTAL', 'WTOTAL']], how='left', on='NONCARB')
    t_df['percent'] = (t_df['_CLLCPWT']*100)/t_df['WTOTAL']
    return t_df
```

```python
column = 'ASTHMA'
df.groupby(['NONCARB', column])[column, '_CLLCPWT'].agg({ column: 'count', '_CLLCPWT': 'sum'}).rename(columns={column: 'count'}).reset_index()
```
